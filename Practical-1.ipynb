```python
import os
import time

board = [' ',' ',' ',' ',' ',' ',' ',' ',' ',' ']
player = 1

win= [
    [1,2,3], [4,5,6], [7,8,9],
    [1,4,7], [2,5,8], [3,6,9],
    [1,5,9], [3,5,7]
]
Win = 1
Draw = -1
Running = 0
Stop = 1

Game = Running
Mark = 'X'

#This Function Draws Game Board
def DrawBoard():
    print(" %c | %c | %c " % (board[1],board[2],board[3]))
    print("___|___|___")
    print(" %c | %c | %c " % (board[4],board[5],board[6]))
    print("___|___|___")
    print(" %c | %c | %c " % (board[7],board[8],board[9]))
    print("   |   | ")

#check if position is empty
def CheckPosition(x):
    if(board[x] == ' '):
        return True
    else:
        return False 
    
def CheckWin():
    global Game
    if(board[1]!=' ' and board[2]!=' ' and board[3]!=' ' and board[4]!=' ' and board[5]!=' ' and board[6]!=' ' and board[7]!=' ' and board[8]!=' ' and board[9]!=' '):
        Game=Draw
    else:
        for a,b,c in win:
            #print(a)
            #print(b)
            #print(c)
            #print("-")
            if(board[a]!=' ' and board[a]==board[b] and board[b]==board[c] ):
                Game = Win
                #print("game : "+str(Game))
                return

print("Player 1: X\nPlayer 2: O\n")
print()
print()

while(Game == Running):
    os.system('cls')
    DrawBoard()
    if(player % 2 != 0):
        print("Player 1's chance")
        Mark = 'X'
    else:
        print("Player 2's chance")
        Mark = 'O'
    choice = int(input())
    if(CheckPosition(choice)):
        board[choice] = Mark
        player=player+1
        CheckWin()
        #print("game"+str(Game))
        if(Game==Win):
            break

os.system('cls')
DrawBoard()
if(Game==Draw):
    print("Game Draw")
elif(Game==Win):
    player-=1
    if(player%2!=0):
        print("Player 1 Won")
    else:
        print("Player 2 Won") 
```
